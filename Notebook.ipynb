```python
import spav

import bokeh
from bokeh.io import output_notebook, show

# suppress the warning caused by the "empty" colorbar figures
bokeh.core.validation.silence(bokeh.core.validation.warnings.MISSING_RENDERERS,True)

output_notebook()
```

### Define the directories

```python
data_filename = 'data/data.hdf5'
static_directory = 'static'
```

### Multiple arrays

```python
def modify_doc(doc):
    view = spav.ExpressionOnArrays(data_filename,static_directory)
    doc.add_root(view.layout)
    
show(modify_doc,notebook_url='localhost:8889')
```

### Single array

```python
def modify_doc(doc):
    view = spav.ExpressionOnArray(data_filename,static_directory)
    doc.add_root(view.layout)
    
show(modify_doc,notebook_url='localhost:8889')
```

### Common coordinate

```python
def modify_doc(doc):
    view = spav.ExpressionInCommonCoordinate(data_filename)
    doc.add_root(view.layout)
    
show(modify_doc,notebook_url='localhost:8889')
```

### Coefficients grouped by AAR

```python
def modify_doc(doc):
    view = spav.LevelExpressionCoefficients(data_filename)
    doc.add_root(view.layout)
    
show(modify_doc,notebook_url='localhost:8889')
```

### Coefficients grouped by the level 1 variables

```python
def modify_doc(doc):
    view = spav.AARExpressionCoefficients(data_filename)
    doc.add_root(view.layout)
    
show(modify_doc,notebook_url='localhost:8889')
```
